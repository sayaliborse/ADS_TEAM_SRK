```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
from plotly import tools
from plotly.offline import iplot, init_notebook_mode
init_notebook_mode()
```

```python
terror_data = pd.read_csv('C:/Users/sayal/OneDrive/Documents/ADS_SAYALI/Homework 1/globalterrorismdb_0718dist.csv',
                          encoding='ISO-8859-1',
                          usecols=[0, 1, 2, 3, 7, 8,10,11, 27, 29, 35, 39, 58,64,82,98, 100,101, 103])
```

```python
terror_data.rename(columns={'eventid':'id', 'iyear':'Year', 'imonth':'Month', 'iday':'Day',
             'country_txt':'Country', 'provstate':'State', 'targtype1_txt':'Target_Type',
             'weaptype1_txt':'Weapon', 'nkillter':'Fatalities', 'nwoundte':'Injuries',
             'country': 'Countryid' , 'region_txt':'Region',
                            'attacktype1': 'attacktype_id', 'attacktype1_txt':'AttackType','target1':'Target',
                            'nkill':'Killed','nwound':'Wounded','gname':'Group',
                            'motive':'Motive' , 'suicide' : 'Suicide'} ,inplace=True)
```

```python
terror_data['Fatalities'] = terror_data['Fatalities'].fillna(0).astype(int)
```

```python
terror_data['Injuries'] = terror_data['Injuries'].fillna(0).astype(int)
```

```python
terror_usa = terror_usa.sort_values(['fatalities', 'injuries'], ascending = False)
```

```python
terror_usa = terror_usa.drop_duplicates(['date', 'latitude', 'longitude', 'fatalities'])
```

```python
bar_graph = terror_data.groupby('Year')['Killed'].size()
plt.rcParams['figure.figsize'] = (20,10)

plt.xlabel('Year of Attack', fontsize=25)
plt.ylabel('No. of Kills', fontsize = 25)
plt.title('No of Kills basis year', fontsize = 30)
bar_graph.plot.bar()
```

```python
suicide_len=len(terror_data[terror_data['Suicide']== 1].index)
suicide_len
nosuicide_len = len(terror_data[terror_data['Suicide']== 0].index)
slices = (suicide_len, nosuicide_len)
plt.rcParams['figure.figsize'] = (20,10)

label = ('Suicide', 'Non-Suicide')
plt. pie(slices, labels=label, autopct='%1.0f%%')
```

```python
terror_usa['text'] = terror_usa['date'].dt.strftime('%B %-d, %Y') + '<br>' +\
                     terror_usa['fatalities'].astype(str) + ' Killed, ' +\
                     terror_usa['injuries'].astype(str) + ' Injured'
```

```python
fatality = dict(
           type = 'scattergeo',
           locationmode = 'USA-states',
           lon = terror_usa[terror_usa.fatalities > 0]['longitude'],
           lat = terror_usa[terror_usa.fatalities > 0]['latitude'],
           text = terror_usa[terror_usa.fatalities > 0]['text'],
           mode = 'markers',
           name = 'Fatalities',
           hoverinfo = 'text+name',
           marker = dict(
               size = terror_usa[terror_usa.fatalities > 0]['fatalities'] ** 0.255 * 8,
               opacity = 0.95,
               color = 'rgb(240, 140, 45)')
           )
```

```python
injury = dict(
         type = 'scattergeo',
         locationmode = 'USA-states',
         lon = terror_usa[terror_usa.fatalities == 0]['longitude'],
         lat = terror_usa[terror_usa.fatalities == 0]['latitude'],
         text = terror_usa[terror_usa.fatalities == 0]['text'],
         mode = 'markers',
         name = 'Injuries',
         hoverinfo = 'text+name',
         marker = dict(
             size = (terror_usa[terror_usa.fatalities == 0]['injuries'] + 1) ** 0.245 * 8,
             opacity = 0.85,
             color = 'rgb(20, 150, 187)')
         )
```

```python
layout = dict(
         title = 'Terrorist Attacks by Latitude/Longitude in United States (1970-2015)',
         showlegend = True,
         legend = dict(
             x = 0.85, y = 0.4
         ),
         geo = dict(
             scope = 'usa',
             projection = dict(type = 'albers usa'),
             showland = True,
             landcolor = 'rgb(250, 250, 250)',
             subunitwidth = 1,
             subunitcolor = 'rgb(217, 217, 217)',
             countrywidth = 1,
             countrycolor = 'rgb(217, 217, 217)',
             showlakes = True,
             lakecolor = 'rgb(255, 255, 255)')
         )
```

```python
data = [fatality, injury]
figure = dict(data = data, layout = layout)
iplot(figure)
```

```python
terror_peryear = np.asarray(terror_usa.groupby('year').year.count())

terror_years = np.arange(1970, 2016)
# terrorist attacks in 1993 missing from database
terror_years = np.delete(terror_years, [23])

trace = [go.Scatter(
         x = terror_years,
         y = terror_peryear,
         mode = 'lines',
         line = dict(
             color = 'rgb(240, 140, 45)',
             width = 3)
         )]

layout = go.Layout(
         title = 'Terrorist Attacks by Year in United States (1970-2015)',
         xaxis = dict(
             rangeslider = dict(thickness = 0.05),
             showline = True,
             showgrid = False
         ),
         yaxis = dict(
             range = [0.1, 425],
             showline = True,
             showgrid = False)
         )

figure = dict(data = trace, layout = layout)
iplot(figure)
```
