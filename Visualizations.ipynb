```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1', low_memory=False)
bar_graph = data.groupby('iyear')['nkill'].size()
plt.rcParams['figure.figsize'] = (20,10)

plt.xlabel('Year of Attack', fontsize=25)
plt.ylabel('No. of Kills', fontsize = 25)
plt.title('No of Kills basis year', fontsize = 30)
bar_graph.plot.bar()
```
